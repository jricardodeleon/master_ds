Codificación de variables categóricas
Codifique la siguiente variable categórica usando el método de codificación Ficticia. Incluya una columna de NaN para valores perdidos.

```python
import pandas as pd
import numpy as np

aceros=['Corten','Corrugado','Galvanizado','Inoxidable','al Carbón']
precios=[(650,800),(600,700),(250,350),(2000,2500),(6000,6500)]

np.random.seed(123)
categórica=np.random.choice(aceros,size=50)
numérica=[]
for c in categórica:
  numérica.append(np.random.randint(*precios[aceros.index(c)]))
df=pd.DataFrame()
df['Precio']=numérica
df['Acero']=categórica

nans=np.random.randint(50,size=np.random.randint(5,10))
df.iloc[nans]=np.NaN
```

```python
df.head(5)
```

```python
# Códificando la variable categorica "Acero" con el metodo de códficacion ficticia

df_dum = pd.get_dummies(df, dummy_na=True, prefix='', prefix_sep="")
df_dum.head(5)
```

```python
# unimos con el data frame original

df_with_dum = pd.concat([df, df_dum], axis=1)
df_with_dum.head(5)
```
